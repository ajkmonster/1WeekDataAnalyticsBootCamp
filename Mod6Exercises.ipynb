```python
df.corr()
```

### Exercise 2:

Using the dataframe from the previous exercise, choose features (columns) to create a linear regression formula to predict sales. Try it with and without the y-intercept. How does it make a difference? Does adding or removing features in your model formula make a difference in the output?

```python
import statsmodels.formula.api as smf
```

```python
result = smf.ols('Sales ~ Location + PeopleNearby + RestaurantsNearby + FoodCartsNearby + Price', data=df).fit()
```

```python
result.summary()
```

```python
result2 = smf.ols('Sales ~ PeopleNearby + RestaurantsNearby + FoodCartsNearby + Price', data=df).fit()
result2.summary()
```

```python
result3 = smf.ols('Sales ~ PeopleNearby + FoodCartsNearby + Price', data=df).fit()
result3.summary()
```

```python
result3 = smf.ols('Sales ~ PeopleNearby + FoodCartsNearby + Price - 1', data=df).fit()
result3.summary()
```

```python
result4 = smf.ols('Sales ~ PeopleNearby + Price - 1', data=df).fit()
result4.summary()
```
